# Data preparation for Country wise analysis

```python
x=netflix_tv.dropna(inplace=False).to_numpy()
```

```python
def strip(x):
    if x == 'USA':
        x='united states'
    elif x == 'UK':
        x='united kingdom'
    
    return x.strip().lower()
```

```python
def change(x):
    y=[]
    for i in range(len(x)):
        y.extend(list(map(strip,x[i].split(','))))
    return y
```

```python
def count_c(x):
    d=dict()
    y=change(x)
    for v in y:
        if v in d.keys():
            d[v]+=1
        else:
            d[v]=1
    return d
```

```python
def mal(x,c_i,t_i,r_i):
    data={"count":0}
    country={"india":data}
    data2={"count":1}
    for v in x:
        t=list(map(strip,v[c_i].split(',')))
        for v2 in t:
            if v2 in country.keys():
                country[v2]['count']+=1
                if v[t_i].lower() in country[v2].keys():
                    country[v2][v[t_i].lower()]+=1
                else:
                    country[v2][v[t_i].lower()]=1
                if v[r_i].lower() in country[v2].keys():
                    country[v2][v[r_i].lower()]+=1
                else:
                    country[v2][v[r_i].lower()]=1
            else:
                country[v2]=data2.copy()
                #print(country)
                if v[t_i].lower() in country[v2].keys():
                    country[v2][v[t_i].lower()]+=1
                else:
                    country[v2][v[t_i].lower()]=1
                if v[r_i].lower() in country[v2].keys():
                    country[v2][v[r_i].lower()]+=1
                else:
                    country[v2][v[r_i].lower()]=1
    
    return country
```
